## Importaciones necesarias.

```python
import os
import spotipy
import json
import pandas as pd

from spotipy.oauth2 import SpotifyOAuth
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id="4480efffdfe540a19c709895e88ea143",
    client_secret="b237bee5eb354fdeb7e907c299a37a15",
    redirect_uri="http://127.0.0.1:8000/callback"
))
```

```python
artista_json = sp.search(q="artist:Oasis")
```

```python
print(json.dumps(artista_json, indent=4))
```

```python
artista_json
```

```python
print(len(artista_json["tracks"]["items"]))
```

```python
id_artista = artista_json["tracks"]["items"][0]["artists"][0]["id"]
nombre_artista = artista_json["tracks"]["items"][0]["artists"][0]["name"]
```

```python
id_artista
```

```python
albums = sp.artist_albums(id_artista, album_type="album", limit = 50)
```

```python
print(len(albums["items"]))
```

```python
lista_albums = []

for i in albums["items"]:
   id_album = i["id"]
   name_album = i["name"]
   canciones = sp.album_tracks(id_album)
   for j in canciones["items"]:
      lista_albums.append({
        "nombre_artsita": nombre_artista,
        "nombre_disco":name_album,
        "nombre_cancion": j["name"]
      })

```

```python
print(len(lista_albums))
```

```python
df = pd.DataFrame(lista_albums)
df
```

```python
df.to_csv(f"canciones_{nombre_artista}.csv")
```
