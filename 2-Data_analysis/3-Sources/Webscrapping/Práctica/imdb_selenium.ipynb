# Webscraping con Selenium de IMDB

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd

service = Service(executable_path='./chromedriver.exe')
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
#time.sleep(5)
#driver.quit()
url = "https://www.imdb.com/chart/top/"
driver.get(url)
time.sleep(1)
boton = driver.find_element(By.CSS_SELECTOR, 'button[data-testid="accept-button"]')
boton.click()

nombre = []
anio = []
duracion = []
rating = []

contenedores = driver.find_elements(By.CSS_SELECTOR, "div.cli-children")
```

```python
for pelicula in contenedores:
    titulo = pelicula.find_element(By.CSS_SELECTOR, "h3.ipc-title__text--reduced")
    nombre.append(titulo.text)
    elementos = pelicula.find_elements(By.CSS_SELECTOR, "span.cli-title-metadata-item")
    anio.append(elementos[0].text)
    duracion.append(elementos[1].text)
    punt = pelicula.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--rating")
    rating.append(punt.text)
```

```python
dic = {
    "titulo" : nombre,
    "año" : anio,
    "duracion" : duracion,
    "rating": rating
}

df = pd.DataFrame(dic)
```
