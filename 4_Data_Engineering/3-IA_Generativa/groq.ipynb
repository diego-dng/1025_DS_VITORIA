# Ejemplos:

## Definicion de las variables necesarias.

```python
# APIKEY
api_key = ""
```

## Prueba con Groq Cloud

### Roles en Groq (Chat Completions)

### system
- Define el comportamiento del modelo
- Reglas, estilo y contexto global
- “Instrucciones internas”

### user
- Input del usuario final
- Preguntas o instrucciones

### assistant
- Respuesta generada por el modelo
- Puede usarse para dar ejemplos previos (few-shot)

### tool
- Resultado de herramientas o funciones externas
- Usado en agents, RAG y automatización


```python
from groq import Groq

client = Groq(api_key=api_key)

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": "Eres un profesor de data science."},
        {"role": "user", "content": "Explica qué es RAG en una frase sencilla"}
    ]
)

print(response.choices[0].message.content)
```
