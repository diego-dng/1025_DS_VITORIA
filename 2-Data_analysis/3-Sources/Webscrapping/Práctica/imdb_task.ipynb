## IMPORTS NECESARIOS

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from fake_useragent import UserAgent
```

## CODIGO

```python
url = "https://www.imdb.com/chart/top/"
ua = UserAgent()
headers = {'User-Agent': ua.random}
response = requests.get(url, headers=headers)
```

```python
response
```

```python
sp = bs(response.content, "html.parser")
```

```python
sp
```

```python
titulos = []
for i in sp.find_all("h3", class_ = "ipc-title__text"):
    titulos.append(i.text)
```

```python
titulos
```

```python
anio = []
duracion = []
for j in sp.find_all("span", class_ = "cli-title-metadata-item"):
    if len(j.text) == 4:
        print(j.text)
        anio.append(j.text)
    elif "h" in j.text:
        duracion.append(j.text)
```

```python
print(anio)
print(duracion)
```

```python
print(len(anio))
print(len(titulos))
```

```python
rating = []
for i in sp.find_all("span", class_ = "ipc-rating-star--rating"):
    rating.append(i.text)
```

```python
print(rating)
```

```python
print(len(titulos))
print(len(anio))
print(len(duracion))
print(len(rating))
```

```python
titulos = titulos[:-1]
print(len(titulos))
```

```python
dic = {
    "titulo" : titulos,
    "año" : anio,
    "duracion" : duracion,
    "rating": rating
}

df = pd.DataFrame(dic)
```

```python
df
```

## JSON

```python
import json
```

```python
data = json.loads(sp.find("script", {"type":"application/ld+json"}).text)
```

```python
data
```

```python
print(len(data["itemListElement"]))
```

```python
data["itemListElement"][4]
```

```python
l_titulo = []
l_anio = []
l_duracion  = []
l_rating = []
for i in data["itemListElement"]:
    l_titulo.append(i["item"].get("alternateName", i["item"]["name"]))
    l_duracion.append(i["item"]["duration"][2:])
    l_rating.append(i["item"]["aggregateRating"]["ratingValue"])
   
        
```

```python
print(len(l_titulo))
print(len(l_duracion))
print(len(l_rating))
```

```python
data = json.loads(sp.find("script", {"id":"__NEXT_DATA__"}).text)
```

```python
data["props"]["pageProps"]["pageData"]["chartTitles"]["edges"]
```

```python
l_posicion = []
for i in data["props"]["pageProps"]["pageData"]["chartTitles"]["edges"]:
    l_posicion.append(i["currentRank"])
    l_anio.append(i["node"]["releaseYear"]["year"])
```

```python
print(len(l_posicion))
print(len(l_titulo))
print(len(l_duracion))
print(len(l_rating))
print(len(l_anio))
```

```python
dic = {
    "l_posicion":l_posicion,
    "titulo" : l_titulo,
    "año" : l_anio,
    "duracion" : l_duracion,
    "rating": l_rating
}

df = pd.DataFrame(dic)
```

```python
df
```

```python
df.to_csv("top_250.csv")
```
