# Conexión a una base de datos PostGresSql

```python
import psycopg2
import pandas as pd
```

```python
def conexion(hosting, puerto, nombre_db, usuario, pswd):
        conn = psycopg2.connect(
                host = hosting,
                port = puerto,
                dbname = nombre_db,
                user = usuario,
                password = pswd
        )
        cursor = conn.cursor()
        return conn, cursor

def sql_query(cursor, query):   # SELECT * FROM TRACKS

    # Ejecuta la query
    cursor.execute(query)

    # Almacena los datos de la query 
    ans = cursor.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in cursor.description]

    return pd.DataFrame(ans,columns=names)

def sql_insert(cursor, conn,query):
    try:
        cursor.execute(query)
        conn.commit()
        return "OK"
    except Exception as e:
        return f"error: {e}" 
```

```python
conexion, cursor = conexion(hosting, puerto, nombre_db, usuario, pswd)
```

```python
query_libro = "SELECT * FROM libro"
query_autor = "SELECT * FROM autor"
insert_autor = "INSERT INTO autor VALUES (3, 'Orwell', 'Britanico')"
insert_libro = "INSERT INTO libro VALUES (4, '1984', 3)"
```
